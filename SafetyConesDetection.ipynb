```
# Model training
from imageai.Detection.Custom import DetectionModelTrainer
import os

execution_path = os.getcwd()

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(os.path.join(execution_path, "drive/", "My Drive/", "traffic_cones_1/"))
trainer.setTrainConfig(object_names_array=["trafficcone"], 
                       batch_size=4, 
                       num_experiments=100, 
                       train_from_pretrained_model="pretrained-yolov3.h5")
trainer.trainModel()
```

```
# Evaluating 7 best (min loss) models obtained as result of training
from imageai.Detection.Custom import DetectionModelTrainer
import os

execution_path = os.getcwd()

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(os.path.join(execution_path, "drive/", "My Drive/", "traffic_cones_1/"))
trainer.evaluateModel(model_path=os.path.join(execution_path, "drive/", "My Drive/", "traffic_cones_1/", "models/"), 
                      json_path=os.path.join(execution_path, "drive/", "My Drive/", "traffic_cones_1/", "json/", "detection_config.json"), 
                      iou_threshold=0.5, 
                      object_threshold=0.3, 
                      nms_threshold=0.5)
```

```
# Test of the best model (4.19% loss) on image
from imageai.Detection.Custom import CustomObjectDetection
import os

execution_path = os.getcwd()

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(os.path.join(execution_path, "drive/", "My Drive/", "traffic_cones_1/", "models/",  "detection_model-ex-43--loss-4.19.h5")) 
detector.setJsonPath("drive/My Drive/traffic_cones_1/json/detection_config.json")
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image=os.path.join(execution_path, "drive/", "My Drive/", "traffic_cones_1/", "test_cone_2.jpg"), 
                                             output_image_path=os.path.join(execution_path, "drive/", "My Drive/", "traffic_cones_1/", "cone_found.jpeg"))
for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
```

```
# Cones detection on video
from imageai.Detection.Custom import CustomVideoObjectDetection

import os

execution_path = os.getcwd()

video_detector = CustomVideoObjectDetection()
video_detector.setModelTypeAsYOLOv3()
video_detector.setModelPath(os.path.join(execution_path, "drive/", "My Drive/", "traffic_cones_1/", "models/",  "detection_model-ex-43--loss-4.19.h5"))
video_detector.setJsonPath("drive/My Drive/traffic_cones_1/json/detection_config.json")
video_detector.loadModel()
print("Model loaded. Proceed to video detection")
video_detector.detectObjectsFromVideo(input_file_path=os.path.join(execution_path, "drive/", "My Drive/", "traffic_cones_1/",  "Traffic_cones_ads.mp4"),
                                        output_file_path=os.path.join(execution_path, "drive/", "My Drive/", "traffic_cones_1/",  "Traffic_cones_detected.mp4"), 
                                        frames_per_second=30,
                                        minimum_percentage_probability=40,
                                        log_progress=True)
```
